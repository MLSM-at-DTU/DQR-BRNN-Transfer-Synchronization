```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.array([[.45, .45], [.45, .55]])
```

```python
transfer_delta = .02
schedule_1 = np.array([[.15, .20], [.5-transfer_delta, .45], [.5-transfer_delta, .55], [.85, .80]])
schedule_2 = np.array([[.85, .20], [.5+transfer_delta, .45], [.5+transfer_delta, .55], [.15, .80]])

d_1 = 0.07
real_time_1 = np.array([[.15, .10], [.5-transfer_delta, .37], [.5-transfer_delta, .55+d_1], [.85, .80+d_1]])
real_time_2 = np.array([[.85, .20+d_1], [.5+transfer_delta, .45+d_1], [.5+transfer_delta, .55+d_1], [.15, .80+d_1]])

d_1 = 0.17
d_2 = 0.04
failed_real_time_1 = np.array([[.15, .10], [.5-transfer_delta, .37], [.5-transfer_delta, .55+d_2], [.85, .80+d_2]])
failed_real_time_2 = np.array([[.85, .20+d_1], [.5+transfer_delta, .45+d_1], [.5+transfer_delta, .55+d_1], [.15, .80+d_1]])


fig, ax = plt.subplots(figsize=(16, 4), ncols=3, sharey=True)
ax[0].plot(schedule_1[:, 0], schedule_1[:, 1], marker='o', color='C1', linestyle='--', label='Timetable, Bus 1')
ax[0].plot(schedule_2[:, 0], schedule_2[:, 1], marker='o', color='C2', linestyle='--', label='Timetable, Bus 2')
ax[0].set_xlim(0, 1)
ax[0].set_ylim(0, 1)
ax[0].legend(loc='lower center')
ax[0].set_title('Scheduled Connection')
ax[0].yaxis.set_ticks([])
ax[0].set_ylabel('Time')
ax[0].xaxis.set_ticks([.15, .5, .85])
ax[0].xaxis.set_ticklabels(['Previous/next\nstop', 'Transfer\nsite', 'Next/previous\nstop'])

ax[1].plot(real_time_1[:, 0], real_time_1[:, 1], marker='o', color='C1', label='Observed, Bus 1')
ax[1].plot(schedule_1[:, 0], schedule_1[:, 1], marker='o', color='C1', linestyle='--', alpha=.25)
ax[1].plot(real_time_2[:, 0], real_time_2[:, 1], marker='o', color='C2', label='Observed, Bus 2')
ax[1].plot(schedule_2[:, 0], schedule_2[:, 1], marker='o', color='C2', linestyle='--', alpha=.25)
ax[1].set_xlim(0, 1)
ax[1].set_ylim(0, 1)
ax[1].legend(loc='lower center')
ax[1].set_title('Successfull, but delayed connection')
ax[1].xaxis.set_ticks([.15, .5, .85])
ax[1].xaxis.set_ticklabels(['Previous/next\nstop', 'Transfer\nsite', 'Next/previous\nstop'])

ax[2].plot(failed_real_time_1[:, 0], failed_real_time_1[:, 1], marker='o', color='C1', label='Observed, Bus 1')
ax[2].plot(schedule_1[:, 0], schedule_1[:, 1], marker='o', color='C1', linestyle='--', alpha=.25)
ax[2].plot(failed_real_time_2[:, 0], failed_real_time_2[:, 1], marker='o', color='C2', label='Observed, Bus 2')
ax[2].plot(schedule_2[:, 0], schedule_2[:, 1], marker='o', color='C2', linestyle='--', alpha=.25)
ax[2].set_xlim(0, 1)
ax[2].set_ylim(0, 1)
ax[2].legend(loc='lower center')
ax[2].set_title('Failed and delayed connection')
ax[2].xaxis.set_ticks([.15, .5, .85])
ax[2].xaxis.set_ticklabels(['Previous/next\nstop', 'Transfer\nsite', 'Next/previous\nstop'])

plt.savefig('transfer.pdf', bbox_inches='tight')
plt.savefig('transfer.png', bbox_inches='tight', dpi=300)
```
